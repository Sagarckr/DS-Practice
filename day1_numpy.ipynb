## Reshape and Ravel

Reshape : Change dimension of an array.<br>
Ravel: Change any dimension of array into 1D array. -------> Flatten

```python
reshape_numbers = numbers.reshape(10,5)
reshape_numbers
```

```python
flat_number = numbers.reshape(50)
flat_number
```

```python
ravel_numbers = np.ravel(reshape_numbers)
ravel_numbers
```

## Array Broadcasting

'''

1. All shape of arrays must be same.
2. Atleast one array's dimension row/column must be 1 and other must be equal.
3. Array also can be of 1*1 matrix form

'''

```python
md1 = np.array([
    [1,2,3],
    [5,6,7],
    [8,9,10]
])

md2 = np.array([
    [3,2,1],
    [6,8,9],
    [3,7,5]
])
```

```python
calc = md1 * md2
calc
```
