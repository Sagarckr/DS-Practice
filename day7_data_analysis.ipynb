```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Dataset/cleaned_data1.csv")
df.head()
```

```python
df.info()
```

```python
# find total employees working in departments.
df['Department'].value_counts()
```

```python
# group by (categorical) + 
    #   aggregate function (categorical, continues,continues,discrete) 
        #   -> used in analysis
# aggregate function only -> used in KPIs (mean, min, max, count, sum)
```

```python
emp_dep = df.groupby('Department')['EmpID'].count().reset_index().rename(
    columns = {
        'EmpID': 'Total_Employees'
    }
).sort_values(
    by = 'Total_Employees', ascending = False, ignore_index = True)
emp_dep
```

```python
# find total salary distributed to each employee by their job position.
```

```python
pos_sal = df.groupby('Position')['Salary'].sum().reset_index()
pos_sal.head()
```

```python
pos_sal['Salary'] = pos_sal['Salary'].map("Rs.{:,.2f}".format)
pos_sal.head()
```

```python
def currencyConverter(dataframe, prefix, *args):
    for col in args:
        dataframe[col] = dataframe[col].map(f"{prefix}""{:,.2f}".format)
```

```python
emp_pos_details = df.groupby('Position').agg(
    Total_Salary = ('Salary', 'sum'),
    Total_Employees = ('EmpID', 'count'),
    Max_Salary = ('Salary', 'max'),
    Min_Salary = ('Salary', 'min'),
    Avg_Salary = ('Salary', 'mean')
).reset_index()
emp_pos_details.columns = emp_pos_details.columns.str.replace('_', ' ')

currencyConverter(emp_pos_details, 'Rs.', 'Total Salary', 'Max Salary',
                  'Min Salary', 'Avg Salary')

emp_pos_details.head()
```
