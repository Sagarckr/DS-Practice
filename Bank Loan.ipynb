# KPIs

## Total Loan Applications

```python
total_loan_applications = df['id'].count()
MTD_total_loan_applications = df[df['month_num'] == MTD_Month]['id'].count()
PMTD_total_loan_applications = df[df['month_num'] == PMTD_Month]['id'].count()
MoM_total_loan_applications = ((MTD_total_loan_applications - PMTD_total_loan_applications)/PMTD_total_loan_applications)*100
```

```python
print(f'Total Loan Applications: {total_loan_applications}')
print(f'MTD Total Loan Applications: {MTD_total_loan_applications}')
print(f'PMTD Total Loan Applications: {PMTD_total_loan_applications}')
print(f'MoM Total Loan Applications: {MoM_total_loan_applications:,.2f}%')
```

## Total Funded Amount

```python
total_funded_amount = df['loan_amount'].sum()
MTD_total_funded_amount = df[df['month_num'] == MTD_Month]['loan_amount'].sum()
PMTD_total_funded_amount = df[df['month_num'] == PMTD_Month]['loan_amount'].sum()
MoM_total_funded_amount = ((MTD_total_funded_amount - PMTD_total_funded_amount)/PMTD_total_funded_amount)*100
```

```python
print(f'Total Funded Amount: Rs.{total_funded_amount:,.2f}')
print(f'MTD Total Funded Amount: Rs.{MTD_total_funded_amount:,.2f}')
print(f'PMTD Total Funded Amount: Rs.{PMTD_total_funded_amount:,.2f}')
print(f'MoM Total Funded Amount: {MoM_total_funded_amount:,.2f}%')
```

## Total Amount Received

```python
total_amount_received = df['total_payment'].sum()
MTD_total_amount_received = df[df['month_num'] == MTD_Month]['total_payment'].sum()
PMTD_total_amount_received = df[df['month_num'] == PMTD_Month]['total_payment'].sum()
MoM_total_amount_received = ((MTD_total_amount_received - PMTD_total_amount_received)/PMTD_total_amount_received)*100
```

```python
print(f'Total Amount Received: Rs.{total_amount_received:,.2f}')
print(f'MTD Total Amount Received: Rs.{MTD_total_amount_received:,.2f}')
print(f'PMTD Total Amount Received: Rs.{PMTD_total_amount_received:,.2f}')
print(f'MoM Total Amount Received: {MoM_total_amount_received:,.2f}%')
```

## Average Interest Rate

```python
average_int_rate = df['int_rate'].mean()
MTD_average_int_rate = df[df['month_num'] == MTD_Month]['int_rate'].mean()
PMTD_average_int_rate = df[df['month_num'] == PMTD_Month]['int_rate'].mean()
MoM_average_int_rate = ((MTD_average_int_rate - PMTD_average_int_rate)/PMTD_average_int_rate)*100
```

```python
print(f'Average Interest Rate: {average_int_rate * 100:,.2f}%')
print(f'MTD Average Interest Rate: {MTD_average_int_rate * 100:,.2f}%')
print(f'PMTD Average Interest Rate: {PMTD_average_int_rate * 100:,.2f}%')
print(f'MoM Average Interest Rate: {MoM_average_int_rate:,.2f}%')
```

## Average Debt-to-Income Ratio

```python
average_dti = df['dti'].mean()
MTD_average_dti = df[df['month_num'] == MTD_Month]['dti'].mean()
PMTD_average_dti = df[df['month_num'] == PMTD_Month]['dti'].mean()
MoM_average_dti = ((MTD_average_dti - PMTD_average_dti)/PMTD_average_dti)*100
```

```python
print(f'Average Debt-to-Income Ratio: {average_dti * 100:,.2f}%')
print(f'MTD Average Debt-to-Income Ratio: {MTD_average_dti * 100:,.2f}%')
print(f'PMTD Average Debt-to-Income Ratio: {PMTD_average_dti * 100:,.2f}%')
print(f'MoM Average Debt-to-Income Ratio: {MoM_average_dti:,.2f}%')
```

## Functions to change structure of data

```python
# function to change currency format
def currencyConvert(dataframe, *args):
    for col in args:
        dataframe[col] = dataframe[col].map("Rs.{:,.2f}".format)

# function to change % format
def pctConvert(dataframe, *args):
    for col in args:
        dataframe[col] = (dataframe[col]*100).map("{:,.2f}%".format)
```

## Loan Status Grid View

```python
ls_grid_view = df.groupby('loan_status').agg(
    Total_Loan_Applications = ('id', 'count'),
    Total_Funded_Amount = ('loan_amount', 'sum'),
    Total_Amount_Received = ('total_payment', 'sum'),
    Average_Interest_Rate = ('int_rate', 'mean'),
    Average_DTI_Rate = ('dti', 'mean')
).reset_index().rename(columns = {
    'loan_status': 'Loan Status', 'Total_Loan_Applications': 'Total Loan Applications',
    'Total_Funded_Amount': 'Total Funded Amount', 'Total_Amount_Received': 'Total Amount Received',
    'Average_Interest_Rate': 'Average Interest Rate', 'Average_DTI_Rate': 'Average DTI Rate'
})
currencyConvert(ls_grid_view, 'Total Funded Amount', 'Total Amount Received')
pctConvert(ls_grid_view, 'Average Interest Rate', 'Average DTI Rate')
ls_grid_view
```

## Analysis

### Monthly Trends by Issue Date:  To identify seasonality and long-term trends in lending activities

```python
monthly_trend = df.groupby(['month_num', 'month_name']).agg(
    Total_Loan_Applications = ('id', 'count'),
    Total_Funded_Amount = ('loan_amount', 'sum'),
    Total_Amount_Received = ('total_payment', 'sum')
).reset_index().rename(columns = {
    'month_name': 'Month Name', 
    'Total_Loan_Applications': 'Total Loan Applications',
    'Total_Funded_Amount': 'Total Funded Amount', 
    'Total_Amount_Received': 'Total Amount Received'
})[
    ['Month Name', 'Total Loan Applications', 'Total Funded Amount', 'Total Amount Received']
  ]

currencyConvert(monthly_trend, 'Total Funded Amount', 'Total Amount Received')
monthly_trend
```

### Regional Analysis by State: To identify regions with significant lending activity and assess regional disparities

```python
regional_analysis = df.groupby('address_state').agg(
    Total_Loan_Applications = ('id', 'count'),
    Total_Funded_Amount = ('loan_amount', 'sum'),
    Total_Amount_Received = ('total_payment', 'sum')
).reset_index().rename(columns = {
    'address_state': 'Address State', 
    'Total_Loan_Applications': 'Total Loan Applications',
    'Total_Funded_Amount': 'Total Funded Amount', 
    'Total_Amount_Received': 'Total Amount Received'
})[
    ['Address State', 'Total Loan Applications', 'Total Funded Amount', 'Total Amount Received']
  ]

currencyConvert(regional_analysis, 'Total Funded Amount', 'Total Amount Received')
regional_analysis
```

### Loan Term Analysis: To allow the client to understand the distribution of loans across various term lengths.

```python
loan_term_analysis = df.groupby('term').agg(
    Total_Loan_Applications = ('id', 'count'),
    Total_Funded_Amount = ('loan_amount', 'sum'),
    Total_Amount_Received = ('total_payment', 'sum')
).reset_index().rename(columns = {
    'term': 'Loan Term', 
    'Total_Loan_Applications': 'Total Loan Applications',
    'Total_Funded_Amount': 'Total Funded Amount', 
    'Total_Amount_Received': 'Total Amount Received'
})[
    ['Loan Term', 'Total Loan Applications', 'Total Funded Amount', 'Total Amount Received']
  ]

currencyConvert(loan_term_analysis, 'Total Funded Amount', 'Total Amount Received')
loan_term_analysis
```

### Employee Length Analysis: How lending metrics are distributed among borrowers with different employment lengths, helping us assess the impact of employment history on loan applications.

```python
employee_length = df.groupby('emp_length').agg(
    Total_Loan_Applications = ('id', 'count'),
    Total_Funded_Amount = ('loan_amount', 'sum'),
    Total_Amount_Received = ('total_payment', 'sum')
).reset_index().rename(columns = {
    'emp_length': 'Employee Length', 
    'Total_Loan_Applications': 'Total Loan Applications',
    'Total_Funded_Amount': 'Total Funded Amount', 
    'Total_Amount_Received': 'Total Amount Received'
})[
    ['Employee Length', 'Total Loan Applications', 'Total Funded Amount', 'Total Amount Received']
  ]

currencyConvert(employee_length, 'Total Funded Amount', 'Total Amount Received')
employee_length
```

### Loan Purpose Breakdown: Will provide a visual breakdown of loan metrics based on the stated purposes of loans, aiding in the understanding of the primary reasons borrowers seek financing.

```python
loan_purpose = df.groupby('purpose').agg(
    Total_Loan_Applications = ('id', 'count'),
    Total_Funded_Amount = ('loan_amount', 'sum'),
    Total_Amount_Received = ('total_payment', 'sum')
).reset_index().rename(columns = {
    'purpose': 'Purpose', 
    'Total_Loan_Applications': 'Total Loan Applications',
    'Total_Funded_Amount': 'Total Funded Amount', 
    'Total_Amount_Received': 'Total Amount Received'
})[
    ['Purpose', 'Total Loan Applications', 'Total Funded Amount', 'Total Amount Received']
  ]

currencyConvert(loan_purpose, 'Total Funded Amount', 'Total Amount Received')
loan_purpose
```

### Home Ownership Analysis: For a hierarchical view of how home ownership impacts loan applications and disbursements.
