## loc and iloc

```python
"""
-loc -> location -> Slice data through alphabetical order
-iloc -> Integer Location -> Slice data through numerical index
"""
```

```python
emp_details.iloc[0:3, 1:4]
```

```python
emp_details.loc['A':'D', 'Age':'Salary']
```

```python
emp_details.iloc[
    [1,4],
    [0,4]
]
```

```python
emp_details.loc[
    ['B', 'E'],
    ['Name', 'Job Position']
]
```
