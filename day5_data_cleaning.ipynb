```python
'''
- Data Wrangling
    - Data Cleaning
        - Fill Null Values:
            - fillna() : used to fill null values.
                        -> if continues Record fill using:
                            mean - if values are in range 
                            median - if values are not in range
                        -> if discrete/categorical fill using:
                            mode, custom record: 0/assume

            - Skewness
            - 0 or near zero -> fill data with mean value
            - if value is less than 0 and equals to or 
              more than 0.5 then fill median value

        - Drop Columns/Rows
            -dropna()
                -> axis -> 0(Row)
                -> axis -> 1(Column)

        - Remove Duplicate Records
            - always do this at first
            - to find duplicate values -> duplicated()
            - to drop duplicate values -> drop_duplicates()

        - Remove unwanted data (special characters / Regular Expressions)
            - str.replace()

        - Removing Whitespaces
            - str.strip() -> remove both side whitespace
            - str.lstrip() -> remove left side whitespace
            - str.rstrip() -> remove right side whitespace

        imp: stats - Mean, Median, Mode, Skewness, Kurtosis, std, variance

    - Feature Engineering


    Average
    -------------
    =Average(Range)
    
    Median
    ---------------
    =Median(Range)
    
    Skewness
    --------------------
    =Skew(Range)
    
    Count Values
    ----------------------
    =countif(range, value)


- Change Data Types:
    -> astype()

'''
```

##  Fill Data

```python
df.isnull().sum()
```

```python
# commission percentage
com_mean = df['COMMISSION_PCT'].mean()
com_median = df['COMMISSION_PCT'].median()

format_mean = f'{com_mean:,.2f}'
format_mean = float(format_mean)

print(com_mean)
print(format_mean)
print(com_median)
```

```python
type(format_mean)
```

```python
df['COMMISSION_PCT'].skew()
```

```python
df['COMMISSON_PCT'] = df['COMMISSION_PCT'].fillna(com_mean)
# df['COMMISSION_PCT].fillna(com_mean, inplace = True)
# df.fillna({'COMMISSION_PCT': com_mean}, inplace = True)
df.head()
```

```python
df.drop(columns = 'COMMISSION_PCT')
df.head()
```

```python
mgr_mode = df['MANAGER_ID'].mode()[0]
dep_mode = df['DEPARTMENT_ID'].mode()[0]
```

```python
print(mgr_mode)
```

```python
print(dep_mode)
```

```python
df.fillna({'MANAGER_ID' : mgr_mode}, inplace = True)
df.fillna({'DEPARTMENT_ID' : dep_mode}, inplace = True)
```
