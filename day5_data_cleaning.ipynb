```python
'''
- Data Wrangling
    - Data Cleaning
        - Fill Null Values:
            - fillna() : used to fill null values.
                        -> if continues Record fill using:
                            mean - if values are in range 
                            median - if values are not in range
                        -> if discrete/categorical fill using:
                            mode, custom record: 0/assume

            - Skewness
            - 0 or near zero -> fill data with mean value
            - if value is less than 0 and equals to or 
              more than 0.5 then fill median value

        - Drop Columns/Rows
            -dropna()
                -> axis -> 0(Row)
                -> axis -> 1(Column)

        - Remove Duplicate Records
            - always do this at first
            - to find duplicate values -> duplicated()
            - to drop duplicate values -> drop_duplicates()

        - Remove unwanted data (special characters / Regular Expressions)
            - str.replace()

        - Removing Whitespaces
            - str.strip() -> remove both side whitespace
            - str.lstrip() -> remove left side whitespace
            - str.rstrip() -> remove right side whitespace

        imp: stats - Mean, Median, Mode, Skewness, Kurtosis, std, variance

    - Feature Engineering


    Average
    -------------
    =Average(Range)
    
    Median
    ---------------
    =Median(Range)
    
    Skewness
    --------------------
    =Skew(Range)
    
    Count Values
    ----------------------
    =countif(range, value)


- Change Data Types:
    -> astype()

'''
```

```python
object_columns.columns
columns = ['Employee_Name', 'EmpID', 'MarriedID', 'MaritalStatusID', 'GenderID',
       'EmpStatusID', 'DeptID', 'PerfScoreID', 'FromDiversityJobFairID',
       'Salary', 'Termd', 'PositionID', 'Position', 'State', 'Zip', 'DOB',
       'Gender', 'MaritalDesc', 'CitizenDesc', 'HispanicLatino', 'RaceDesc',
       'DateofHire', 'DateofTermination', 'TermReason', 'EmploymentStatus',
       'Department', 'ManagerName', 'RecruitmentSource', 'PerformanceScore',
       'EmpSatisfaction', 'SpecialProjectsCount', 'LastPerformanceReview_Date',
       'DaysLateLast30', 'Absences']
```

```python
def finduniquevalues(dataframe, columns):
    for col in columns:
        print(col)
        print('-'*55)
        unique_values = dataframe[col].unique()
        print(unique_values)
        print('\n')

finduniquevalues(hr,columns)
```

```python
hr['Department'] = hr['Department'].str.strip()
```

```python
def removeWhiteSpace(dataframe, columns):
    for col in columns:
        if dataframe[col].dtype == "object" or str(dataframe[col].dtype) == "string":
            dataframe[col] = dataframe[col].str.strip()
    return dataframe
```

```python
data = removeWhiteSpace(hr, columns)
```

```python
data['HispanicLatino'] = data['HispanicLatino'].str.replace('no', 'No')
data['HispanicLatino'] = data['HispanicLatino'].str.capitalize() # convert words first letter to upper case
data['Employee_Name'] = data['Employee_Name'].str.replace(',', ' ', regex=False)
```
